```python
#analise de dados em html - texto
import numpy as np
import pandas as pd
import requests

url = 'https://www.worldcoinindex.com/'
crypto_data = pd.read_html(crypto_url.text)
len(crypto_data), type(crypto_data) #Averiguando informações dos dados obtidos! A partir da saída de len(),fica claro que existe apenas 1 tabela.
crypto_data = crypto_data[0] #Remover a primeira e a segunda coluna, pois elas não contêm nenhuma informação útil e manter todas as linhas
crypto_final = crypto_data.iloc[:,2:]#O indexador iloc é uma função seleciona por números inteiros das linhas, arrays ou por slice

#Trocando o nome de algumas colunas
crypto_final.columns = crypto_final.columns.str.replace('Last price','last_price')
crypto_final.columns = crypto_final.columns.str.replace('24 high', 'high')
crypto_final.columns = crypto_final.columns.str.replace('24 low','low')
crypto_final.columns = crypto_final.columns.str.replace('Price Charts 7d','price_charts')

#convertendo os dados
crypto_data['last_price'] = crypto_data['Last price'].str.replace(',', '').str.replace('.', '').astype(float)
crypto_data['%'] = crypto_data['%'].str.replace('%', '').str.replace(',', '.').astype(float)

#media moda e mediana das colunas last_price e %
media_last_price = crypto_data['last_price'].mean()
mediana_last_price = crypto_data['last_price'].median()
moda_last_price = crypto_data['last_price'].mode().values[0]

media_percent = crypto_data['%'].mean()
mediana_percent = crypto_data['%'].median()
moda_percent = crypto_data['%'].mode().values[0]

print(f"Média de last_price: {media_last_price}")
print(f"Mediana de last_price: {mediana_last_price}")
print(f"Moda de last_price: {moda_last_price}")

print(f"Média de %: {media_percent}")
print(f"Mediana de %: {mediana_percent}")
print(f"Moda de %: {moda_percent}")
```
