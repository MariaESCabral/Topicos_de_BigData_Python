```python
#Praticando An√°lise de Dados - Limpeza dos dados:
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/MariaESCabral/Topicos_de_BigData_Python/main/base_precos.csv', sep = ';')
df.columns= ['ID','DATA','PRODUTO','QUANTIDADE','VALOR UNITARIO','VALOR TOTAL','SETOR'] #Trocar os nomes das coluna
df['SETOR'] = df['SETOR'].str.lower() #Unificar os nomes da coluna 'SETOR'

df['VALOR UNITARIO'] = df['VALOR UNITARIO'].str.strip("R$") #Retirar o cifr√£o ($) das colunas VALOR UNITARIO
df['VALOR TOTAL'] = df['VALOR TOTAL'].str.strip("R$") #Retirar o cifr√£o ($) das colunas VALOR TOTAL

df['VALOR UNITARIO'] = df['VALOR UNITARIO'].str.replace(",", ".") #Trocar as v√≠rgulas por ponto das colunas VALOR UNITARIO
df['VALOR TOTAL'] = df['VALOR UNITARIO'].str.replace(",", ".") #Trocar as v√≠rgulas por ponto das colunas VALOR TOTAL

df['VALOR UNITARIO'] = pd.to_numeric(df['VALOR UNITARIO'], errors='coerce') #Converter para float as colunas VALOR UNITARIO
df['VALOR TOTAL'] = pd.to_numeric(df['VALOR TOTAL'], errors='coerce') #Converter para float as colunas VALOR TOTAL

media1 = df['VALOR TOTAL'].mean()
media2 = df['VALOR UNITARIO'].mean()

mediana1 = df['VALOR TOTAL'].median()
mediana2 = df['VALOR UNITARIO'].median()

moda1 = df['VALOR TOTAL'].mode()
moda2 = df['VALOR UNITARIO'].mode()

desvio1 = df['VALOR TOTAL'].std()
desvio2 = df['VALOR UNITARIO'].std()
```

```python
#REGRESS√ÉO LINEAR
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

x = np.array([9714, 3728, 6062, 8845, 8378]) #variavel independente
y = np.array([3011, 305, 1879 ,2654, 2849]) #variavel dependente

coeficiente_angular, intercepto = np.polyfit(x, y, 1)

print(f"Coeficiente Angular: {coeficiente_angular}") #calcular a regressao linear simples
print(f"Intercepto:{intercepto}")

y_pred = coeficiente_angular * x + intercepto #previsoes usando a equa√ß√£o da regressao
print(y_pred)

plt.scatter(x,y, label= "Dados Originais") #Define um gr√°fico de dispers√£o de ùë¶ vs. ùë• com tamanho e/ou cor de marcador variados
plt.plot(x, y_pred, color='red', linewidth=2, label='linha de regressao') #Plota ùë¶ vs. ùë• como uma linha, cor e marcadores variados

plt.xlabel('variavel independente (X)')#R√≥tulos de de ùë• e ùë¶.
plt.ylabel('variavel dependente (Y)')

plt.legend() #Classe legenda para inserir legendas no seu gr√°fico
plt.title('regressao linear simples com numpy')

plt.grid(True) #Insere linhas de grades no gr√°fico.
plt.show
```

```python
# Obtenha o dataset do endere√ßo abaixo e realize as seguintes an√°lises:
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Criando os dados dos visitantes
dias = np.arange(1, 21)  # Dias de 1 a 20
visitantes = 20 + 20 * (dias - 1)  # Aumento de 20 visitantes por dia

# Transformando os dados em formato de matriz (necess√°rio para o scikit-learn)
dias = dias.reshape(-1, 1)

# Criando o modelo de regress√£o linear
modelo = LinearRegression()

# Ajustando o modelo aos dados
modelo.fit(dias, visitantes)

# Realizando previs√µes com o modelo
previsoes = modelo.predict(dias)

# Plotando o gr√°fico de dispers√£o
plt.scatter(dias, visitantes, label='Dados reais', color='blue')
plt.plot(dias, previsoes, label='Regress√£o Linear', color='red')
plt.xlabel('Dias')
plt.ylabel('N√∫mero de Visitantes')
plt.title('Gr√°fico de Dispers√£o e Regress√£o Linear')
plt.legend()
plt.show()

# Coeficientes da regress√£o linear
coeficiente_linear = modelo.intercept_
coeficiente_inclinacao = modelo.coef_[0]

print(f"Coeficiente de Intercepta√ß√£o: {coeficiente_linear}")
print(f"Coeficiente de Inclina√ß√£o: {coeficiente_inclinacao}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

dias = np.arange(1, 21)
visitantes = 20 + 20 * (dias - 1)

dias = dias.reshape(-1, 1)
modelo = LinearRegression()
modelo.fit(dias, visitantes)
previsoes = modelo.predict(dias)
plt.scatter(dias, visitantes, label='Dados reais', color='blue')
plt.plot(dias, previsoes, label='Regress√£o Linear', color='red')
plt.xlabel('Dias')
plt.ylabel('N√∫mero de Visitantes')
plt.title('Gr√°fico de Dispers√£o e Regress√£o Linear')
plt.legend()
plt.show()
coeficiente_linear = modelo.intercept_
coeficiente_inclinacao = modelo.coef_[0]

print(f"Coeficiente de Intercepta√ß√£o: {coeficiente_linear}")
print(f"Coeficiente de Inclina√ß√£o: {coeficiente_inclinacao}")
```
