```python
#Praticando Análise de Dados - Limpeza dos dados:
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/MariaESCabral/Topicos_de_BigData_Python/main/base_precos.csv', sep = ';')
df.columns= ['ID','DATA','PRODUTO','QUANTIDADE','VALOR UNITARIO','VALOR TOTAL','SETOR'] #Trocar os nomes das coluna
df['SETOR'] = df['SETOR'].str.lower() #Unificar os nomes da coluna 'SETOR'

df['VALOR UNITARIO'] = df['VALOR UNITARIO'].str.strip("R$") #Retirar o cifrão ($) das colunas VALOR UNITARIO
df['VALOR TOTAL'] = df['VALOR TOTAL'].str.strip("R$") #Retirar o cifrão ($) das colunas VALOR TOTAL

df['VALOR UNITARIO'] = df['VALOR UNITARIO'].str.replace(",", ".") #Trocar as vírgulas por ponto das colunas VALOR UNITARIO
df['VALOR TOTAL'] = df['VALOR UNITARIO'].str.replace(",", ".") #Trocar as vírgulas por ponto das colunas VALOR TOTAL

df['VALOR UNITARIO'] = pd.to_numeric(df['VALOR UNITARIO'], errors='coerce') #Converter para float as colunas VALOR UNITARIO
df['VALOR TOTAL'] = pd.to_numeric(df['VALOR TOTAL'], errors='coerce') #Converter para float as colunas VALOR TOTAL

media1 = df['VALOR TOTAL'].mean()
media2 = df['VALOR UNITARIO'].mean()

mediana1 = df['VALOR TOTAL'].median()
mediana2 = df['VALOR UNITARIO'].median()

moda1 = df['VALOR TOTAL'].mode()
moda2 = df['VALOR UNITARIO'].mode()

desvio1 = df['VALOR TOTAL'].std()
desvio2 = df['VALOR UNITARIO'].std()
```

```python
#REGRESSÃO LINEAR
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

x = np.array([9714, 3728, 6062, 8845, 8378]) #variavel independente
y = np.array([3011, 305, 1879 ,2654, 2849]) #variavel dependente

coeficiente_angular, intercepto = np.polyfit(x, y, 1)

print(f"Coeficiente Angular: {coeficiente_angular}") #calcular a regressao linear simples
print(f"Intercepto:{intercepto}")

y_pred = coeficiente_angular * x + intercepto #previsoes usando a equação da regressao
print(y_pred)

plt.scatter(x,y, label= "Dados Originais") #Define um gráfico de dispersão de 𝑦 vs. 𝑥 com tamanho e/ou cor de marcador variados
plt.plot(x, y_pred, color='red', linewidth=2, label='linha de regressao') #Plota 𝑦 vs. 𝑥 como uma linha, cor e marcadores variados

plt.xlabel('variavel independente (X)')#Rótulos de de 𝑥 e 𝑦.
plt.ylabel('variavel dependente (Y)')

plt.legend() #Classe legenda para inserir legendas no seu gráfico
plt.title('regressao linear simples com numpy')

plt.grid(True) #Insere linhas de grades no gráfico.
plt.show
```

```python
# Obtenha o dataset do endereço abaixo e realize as seguintes análises:
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Criando os dados dos visitantes
dias = np.arange(1, 21)  # Dias de 1 a 20
visitantes = 20 + 20 * (dias - 1)  # Aumento de 20 visitantes por dia

# Transformando os dados em formato de matriz (necessário para o scikit-learn)
dias = dias.reshape(-1, 1)

# Criando o modelo de regressão linear
modelo = LinearRegression()

# Ajustando o modelo aos dados
modelo.fit(dias, visitantes)

# Realizando previsões com o modelo
previsoes = modelo.predict(dias)

# Plotando o gráfico de dispersão
plt.scatter(dias, visitantes, label='Dados reais', color='blue')
plt.plot(dias, previsoes, label='Regressão Linear', color='red')
plt.xlabel('Dias')
plt.ylabel('Número de Visitantes')
plt.title('Gráfico de Dispersão e Regressão Linear')
plt.legend()
plt.show()

# Coeficientes da regressão linear
coeficiente_linear = modelo.intercept_
coeficiente_inclinacao = modelo.coef_[0]

print(f"Coeficiente de Interceptação: {coeficiente_linear}")
print(f"Coeficiente de Inclinação: {coeficiente_inclinacao}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

dias = np.arange(1, 21)
visitantes = 20 + 20 * (dias - 1)

dias = dias.reshape(-1, 1)
modelo = LinearRegression()
modelo.fit(dias, visitantes)
previsoes = modelo.predict(dias)
plt.scatter(dias, visitantes, label='Dados reais', color='blue')
plt.plot(dias, previsoes, label='Regressão Linear', color='red')
plt.xlabel('Dias')
plt.ylabel('Número de Visitantes')
plt.title('Gráfico de Dispersão e Regressão Linear')
plt.legend()
plt.show()
coeficiente_linear = modelo.intercept_
coeficiente_inclinacao = modelo.coef_[0]

print(f"Coeficiente de Interceptação: {coeficiente_linear}")
print(f"Coeficiente de Inclinação: {coeficiente_inclinacao}")
```
