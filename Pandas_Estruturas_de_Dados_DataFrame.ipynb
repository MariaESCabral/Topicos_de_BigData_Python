```python
#media moda mediana e desvio padrao
import pandas as pd

dados = pd.DataFrame({'valores': [15,15,12,18,20,4,15,17,20,20]})

media = dados ['valores'].mean()
print(f'Media:{media}')

mediana = dados ['valores'].median()
print(f'mediana:{mediana}')

moda = dados ['valores'].mode()
print(f'moda:{moda}')

desvio = dados ['valores'].std()
print(f'desvio:{desvio}')
```
